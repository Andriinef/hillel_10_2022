## - Setup black, isort, flake8 in GitHub Actions

Создаем папку .github, в которой создаем папку workflows.
В папке workflows создаем файл code_quality.yml с конфигурациее

```python
# название
name: Check code quality
    
# целевые события для запуска
# в данном случае это push
on: [push]

# выполняемые действия
jobs:
  build:
    runs-on: ubuntu-latest
    
    # запускаются по шагам
    # name - название шага
    # в github actions будет выводиться название шага, и процессы
    steps:
      - uses: actions/checkout@v3
      - uses: actions/setup-python@v3
        with:
          python-version: "3.10"
      - name: Install pipenv
        run: python -m pip install --upgrade pip && pip install pipenv
      - name: Install deps
        run: pipenv sync --system
      - name: Run linter
        run: flake8 .
      - name: Run formatters
        run: black --check . && isort --check .
```

В файле 1.ру делаем ошибку

commit "test_02" и push на git. Проверяем git и видим ошибку

Делаем исправления в файле 1.ру и снова заливаем на git

Теперь в git код залит без ошибок

## - Setup pre-commit hooks

Прежде чем запускать хуки, необходимо установить менеджер пакетов.

Использование pip:

pre-commit --version покажет, какую версию используем

Запускаем предварительную установку, чтобы настроить скрипты git hook

Создаем файл pre-commit-config.yaml

```python
repos:
  - repo: local
    hooks:
      - id: black
        name: black
        entry: black --check .
        language: system
        types: [python]
        pass_filenames: false
      - id: flake8
        name: flake8
        entry: flake8 .
        language: system
        types: [python]
        pass_filenames: false
      - id: isort
        name: isort
        entry: isort --check .
        language: system
        types: [python]
        pass_filenames: false
```

Сопоставление хука настраивает, какой хук из репозитория используется, и позволяет настраивать его. Все необязательные ключи получат значения по умолчанию из конфигурации репозитория.

Делаем ошибку в файле 1.ру

Пробуем commit на git, что приводит к ошибке

После поиска ошибок и исправления файла 1.ру, commit и push на git выполняеться без ошибок.
